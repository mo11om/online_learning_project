```python
import numpy as np 
import random 
import select_path
import copy
import env
import matplotlib.pyplot as plt
```

```python
coefficient =[[3,2], [4,1],[2,4],[6,0]]
player_number = 100
path_number = len(coefficient)
gradient_times = 100
learn_rate = 0.0005
T = 100
scale = 1
hindsight_real_diff = []
everage_regret = []
potential_value = []
everage_potencial = []
game = env.congestion_game(coefficient, path_number, player_number)
for i in range(1, T+1) :
    game.update_estimate_strategy(gradient_times, learn_rate, scale)
    game.random_select_cost()
    game.update_strategy(gradient_times, learn_rate, scale)
    game.hindsight()
    hindsight_real_diff.append(game.hindsight_real_diff)
    everage_regret.append(sum(hindsight_real_diff)/i)#30rounds 1~30
    potential_value.append(select_path.potential_function(game.total_path_select, game.cost_func))
    everage_potencial.append(sum(potential_value)/i)
# print(everage_regret)
# print(everage_potencial)
```

```python
plt.plot(range(T), everage_potencial)
plt.show()
plt.plot(range(T), potential_value)
plt.title('potiencial function')
plt.show()
```
